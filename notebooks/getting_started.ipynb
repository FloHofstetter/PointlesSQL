# Getting Started with PointlesSQL

The `PQL` helper bridges Unity Catalog metadata (managed by
[soyuz-catalog](https://github.com/FloHofstetter/soyuz-catalog))
and Delta Lake tables.  You can **read** any registered table as a
Pandas DataFrame, **write** a DataFrame back as a Delta table, and
**list** catalogs, schemas, and tables — all from a notebook cell.

```python
from pointlessql import PQL

pql = PQL()
```

## Listing catalogs

See what catalogs are registered in the Unity Catalog server.

```python
pql.list_catalogs()
```

## Reading a table

Use the three-part name `catalog.schema.table` to read a Delta table
as a Pandas DataFrame.  Replace the placeholder below with a real
table from your catalog.

```python
# Replace with your actual catalog.schema.table name
df = pql.table("catalog.schema.table")
df.head()
```

## Writing a table

Create a DataFrame and write it as a new Delta table.  The table
metadata (columns, types) is registered automatically in the catalog.

```python
import pandas as pd

df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["alice", "bob", "charlie"],
    "score": [9.5, 8.0, 7.3],
})

# Replace with your target catalog.schema.table_name
pql.write_table(df, "catalog.schema.my_new_table")
```
